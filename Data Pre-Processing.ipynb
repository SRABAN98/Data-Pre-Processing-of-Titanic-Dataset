```python
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
train["Gender"] = train["Sex"].apply(getNumber)
```

```python
train.head()
```

```python
del train["Sex"]
train.head()
```

```python
train.isnull().sum()
```

```python
meanS = train[train.Survived==1].Age.mean()
meanS
```

```python
meanNS = train[train.Survived==0].Age.mean()
meanNS
```

```python
train["age"] = np.where(pd.isnull(train.Age) & train["Survived"]==1  ,meanS, train["Age"])
train.head()
```

```python
train.isnull().sum()
```

```python
train.age.fillna(meanNS,inplace=True)
train.head()
```

```python
train.isnull().sum()
```

```python
del train["Age"]
train.head()
```

```python
survivedQ = train[train.Embarked == 'Q'][train.Survived == 1].shape[0]
survivedC = train[train.Embarked == 'C'][train.Survived == 1].shape[0]
survivedS = train[train.Embarked == 'S'][train.Survived == 1].shape[0]
print(survivedQ)
print(survivedC)
print(survivedS)
```

```python
survivedQ = train[train.Embarked == 'Q'][train.Survived == 0].shape[0]
survivedC = train[train.Embarked == 'C'][train.Survived == 0].shape[0]
survivedS = train[train.Embarked == 'S'][train.Survived == 0].shape[0]
print(survivedQ)
print(survivedC)
print(survivedS)
```

```python
train.dropna(inplace=True)
train.head()
```

```python
train.isnull().sum()
```

```python
train.rename(columns={'age':'Age'}, inplace=True)
train.rename(columns={'Gender':'Sex'}, inplace=True)
train.head()
```

```python
def getEmb(str):
    if str == "S":
        return 1
    elif str == 'Q':
        return 2
    else:
        return 3
train["Embark"] = train["Embarked"].apply(getEmb)
train.head()
```

```python
del train["Embarked"]
train.rename(columns={"Embark":"Embarked"}, inplace=True)
train.head()
```

```python
males = (train["Sex"] == 1).sum()
females = (train["Sex"] == 2).sum()
print(males)
print(females)
p = [males, females]
plt.pie(p,labels = ["Male","Female"],colors = ["green","yellow"], explode = (0.05, 0), startangle = 0)  
plt.axis("equal") 
plt.show()
```

```python
MaleS = train[train.Sex == 1][train.Survived == 1].shape[0]
print(MaleS)
MaleNS = train[train.Sex == 1][train.Survived == 0].shape[0]
print(MaleNS)
FemaleS = train[train.Sex == 2][train.Survived == 1].shape[0]
print(FemaleS)
FemaleNS = train[train.Sex == 2][train.Survived == 0].shape[0]
print(FemaleNS)
```

```python
chart = [MaleS,MaleNS,FemaleS,FemaleNS]
colors = ["lightskyblue","yellowgreen","Yellow","Orange"]
labels = ["Survived Male","Not Survived Male","Survived Female","Not Survived Female"]
explode = [0,0.05,0,0.1]
plt.pie(chart,labels = labels,colors = colors,explode = explode,startangle = 100,counterclock = False,autopct = "%.2f%%")
plt.axis("equal")
plt.show()
```
